## Step 1. Loading and meeting data for the first time

The data we will use for this example is the Iris dataset, a classical dataset in machine
learning and statistics. It is included in scikit-learn in the datasets module. We
can load it by calling the load_iris function:

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
print('Iris Dataset keys: \n{}'.format(iris.keys()))
```

```python
print('Iris Dataset targets: \n{}'.format(iris['target_names']), '\n\nIris Dataset features: \n{}'.format(iris['feature_names']))
```

```python
print('Dataset power: {}'.format(iris['data'].shape))
```

```python
iris['data'][:10]
```

```python
import numpy as np
print(np.unique(iris['target']))
```

The meanings of the numbers are given by the target_names array:
**0 means setosa, 1 means versicolor, and 2 means virginica.**

## Step 2. Measuring Success: Training and Testing Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris['data'], iris['target'], random_state = 0)
```

```python
print('Power of training data: {}'.format(X_train.shape))
print('Power of testing data: {}'.format(X_test.shape))
print('Power of training targets: {}'.format(y_train.shape))
print('Power of testing targets: {}'.format(y_test.shape))
```

## Step 3. Preprocessing data

```python
import pandas as pd

data = pd.DataFrame(X_train, columns = iris.feature_names)
data
```

```python
import mglearn
graphs = pd.plotting.scatter_matrix(data, c = y_train, figsize = (20,20), hist_kwds = {'bins': 20}, s = 250, cmap = mglearn.cm3);
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
```

```python
knn.fit(X_train, y_train)
```

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
print('X_new shape: {}'.format(X_new.shape))
```

```python
prediction = knn.predict(X_new)
print('Prediction: {}'.format(prediction), '\n\nPredicted target name: {}'.format(iris['target_names'][prediction]))
```

## Evaluating the model

```python
y_pred = knn.predict(X_test)
print('Test predictions: {}'.format(y_pred))
```

```python
print('Accuracy: {:.2f}'.format(np.mean(y_pred == y_test)))
```
